## Edit CSV Scraped Table

### Read CSV Scraped Table

```python
urls = ['/id/kanal/perbankan/regulasi/surat-edaran-bank-indonesia', '/id/kanal/perbankan/regulasi/undang-undang',
        '/id/kanal/perbankan/regulasi/peraturan-bank-indonesia', '/id/kanal/perbankan/Regulasi/peraturan-ojk',
        '/id/kanal/perbankan/Regulasi/surat-edaran-ojk', '/id/kanal/syariah/regulasi/peraturan-perbankan-syariah-pbi-dan-sebi',
        '/id/kanal/syariah/regulasi/undang-undang'
        ]

df_table = pd.read_csv('./log/ojk_table_scraping_result_new.csv')
```

### Add Web URL

```python
# split df['url'] and get first 5
df_changed = df_table
df_changed['Web URL'] = 'https://ojk.go.id' + df_changed['URL'].str.split('/').str[:6].str.join('/') + '/default.aspx'
```

### Add Nomor Regulasi using RegEx

```python
# from utils.new_table_scraper import extract_nomor_regulasi

def extract_nomor_regulasi(title):
    patterns = [
        r"\b\d{1,2}\/\d{1,2}\/\w+(?:\/\d{4})?", 
        r"\b\d{1,2}\/[\w\.]+\/\d{4}",
        r"\b\d{1,2}\sTahun\s\d{4}"
        # r"\b\d{1,2}\/\d{1,2}\w+"
    ]
    
    for pattern in patterns:
        match = re.search(pattern, title)
        print(match)
        if match:
            return match.group()
    return "Nomor regulasi not found"

df_changed['Nomor Regulasi'] = df_changed['Title'].apply(extract_nomor_regulasi)

# check where nomor regulasi is none
df_changed[df_changed['Nomor Regulasi'] == 'Nomor regulasi not found']
```
