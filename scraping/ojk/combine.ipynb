```python
# check if df_sanitized_new 'nomor_regulasi' not end with year 'YYYY' then add last 4 digit from column 'tanggal_berlaku' to 'nomor_regulasi'
import re

year_pattern = re.compile(r'\d{4}$')


for index, row in df_sanitized_new.iterrows():
    if not year_pattern.search(row['nomor_regulasi']):
        year = year_pattern.search(row['tanggal_berlaku']).group()
        df_sanitized_new.at[index, 'nomor_regulasi'] = row['nomor_regulasi'] + "/" + year

df_sanitized_new


# nomor_regulasi = "25/SEOJK.03/2016"

# # check if nomor_regulasi match the pattern

# print(year_pattern.search(nomor_regulasi))
```

```python
month_mapping = {
        'Januari': 'Januari',
        'Jan': 'Januari',
        'Februari': 'Februari',
        'Feb': 'Februari',
        'Maret': 'Maret',
        'Mar': 'Maret',
        'April': 'April',
        'Apr': 'April',
        'Mei': 'Mei',
        'May': 'Mei',
        'Juni': 'Juni',
        'Jun': 'Juni',
        'Juli': 'Juli',
        'Jul': 'Juli',
        'Juliy': 'Juli',
        'Agustus': 'Agustus',
        'Aug': 'Agustus',
        'September': 'September',
        'Sep': 'September',
        'Oktober': 'Oktober',
        'Oct': 'Oktober',
        'November': 'November',
        'Nov': 'November',
        'Desember': 'Desember',
        'Dec': 'Desember',
        'Desemberember': 'Desember',
        'Junii': 'Juni',
        'Maretet': 'Maret',
        'Septembertember': 'September',
        'Januariuari': 'Januari',
        'Februariruari': 'Februari',
        'Aprilil': 'April',
        'Julii': 'Juli',
    }


# change month name to month number from column 'tanggal_berlaku'
for index, row in df_sanitized_new.iterrows():
    for key in month_mapping:
        pattern = re.compile(r'\b' + re.escape(key) + r'\b')
        if pattern.search(row['tanggal_berlaku']):
            df_sanitized_new.at[index, 'tanggal_berlaku'] = pattern.sub(month_mapping[key], row['tanggal_berlaku'])
            break

df_sanitized_new
```

```python
df_sanitized_new.to_csv('./log/testing.csv', index=False)
```

```python
# combine df
combine_df = pd.concat([df_sanitized, df_sanitized_new], ignore_index=True)
combine_df
```

```python
# remove row with column 'new_filename' end with zip and rar
combine_df = combine_df[~combine_df['new_filename'].str.endswith('.zip')]
combine_df = combine_df[~combine_df['new_filename'].str.endswith('.rar')]
combine_df
```

```python
combine_df = combine_df.drop_duplicates(keep='first')
combine_df
```

```python
# remove column 'is_ocr'
combine_df = combine_df.drop(columns=['is_ocr'])
combine_df
```

```python
# if column 'nomor_regulasi' contains '/' remove space
def remove_space(row):
    if '/' in row:
        return row.replace(' ', '')
    return row

combine_df['nomor_regulasi'] = combine_df['nomor_regulasi'].apply(remove_space)
```

```python
combine_df = combine_df.fillna('NO DATA')
combine_df
```

```python
# add id to combine_df
combine_df['doc_id'] = range(1, len(combine_df) + 1)

combine_df.to_csv('./log/combine-all-ojk-web.csv', index=False)
```
